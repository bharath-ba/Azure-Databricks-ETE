```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
dbutils.widgets.text("init_load_flag", "0")  # default value is "0"
init_load_flag = int(dbutils.widgets.get("init_load_flag"))
```

# Data Reading

```python
df = spark.sql("select * from databricks_cata.silver.customers_silvers")
```

# Removing Duplicates

```python
df = df.dropDuplicates(subset=['customer_id'])
```

# dividing new vs Old Records

```python
 if init_load_flag == 0:
    df_old = spark.sql('''select DimCustomersKey, customer_id, create_date, update_date
                        from databricks_cata.gold.DimCustomers''')
else:
    df_old = spark.sql('''select 0 DimCustomersKey, 0 customer_id, 0 create_date, 0 update_date
                        from databricks_cata.silver.customers_silvers where 1=0''')
```

```python
df_old.display()
```

### renaming columns of df_old

```python
    df_old = df_old.withColumnRenamed("DimCustomersKey", "old_DimCustomerKey")\
        .withColumnRenamed("customer_id", "old_customer_id")\
        .withColumnRenamed("create_date", "old_create_date")\
        .withColumnRenamed("update_date", "old_update_date")
```

## Applying join with old records

```python
df_join = df.join(df_old, df.customer_id == df_old.old_customer_id, "left")
df_join.limit(5).display()
```

### sepearting new v/s old value

```python
df_new = df_join.filter(df_join['old_DimCustomerKey'].isNull())
```

```python
df_old = df_join.filter(df_join['old_DimCustomerKey'].isNotNull())
```

### preparing df_old

```python
# Dropping all the columns not required
df_old = df_old.drop('old_customer_id','old_update_date')

# Renaming "old_created_date" to "create_date" column
df_old = df_old.withColumnRenamed("old_DimCustomerKey", "DimCustomersKey")

# Renaming "old_created_date" to "create_date" column
df_old = df_old.withColumnRenamed("old_create_date", "create_date")
df_old = df_old.withColumn("create_date",to_timestamp(col("create_date")))

# Recreating "update_date" column with current timestamp
df_old = df_old.withColumn("update_date", current_timestamp())

```

```python
df_old.display()
```

### preparinf df new

```python
# Dropping all the columns not required
df_new = df_new.drop('old_DimCustomerKey','old_customer_id','old_create_date','old_update_date')

# Renaming "dreated_date" to "update_date" columns with timestamp
df_new = df_new.withColumn("update_date", current_timestamp())
df_new = df_new.withColumn("create_date",current_timestamp())

```

```python
df_new.display()
```

# Surrogate key - From 1

```python
df_new = df_new.withColumn("DimCustomersKey",monotonically_increasing_id()+lit(1))
```

```python
df_old.display()
```

# Add Max Surrogate Key

```python
if init_load_flag == 1:
    max_surrogate_key = 0
else:
    df_max_sur = spark.sql('''select max(DimCustomersKey) as max_surrogate_key
                        from databricks_cata.gold.DimCustomers''')
    #converting df_maxsur to max_surrogate_key variable
    max_surrogate_key = df_max_sur.collect()[0]['max_surrogate_key']
    
    
```

```python
df_new = df_new.withColumn("DimCustomersKey",lit(max_surrogate_key+col("DimCustomersKey")))
```

```python
df_old.display()
```

### union of df_old and df_new

```python
df_final = df_new.unionByName(df_old)
```

```python
df_final.display()
```

# SCD 1

```python
from delta.tables import DeltaTable 
```

```python
if spark.catalog.tableExists("databricks_cata.gold.DimCustomers"):
    dlt_obj = DeltaTable.forPath(spark, "abfss://gold@databrickssabharath.dfs.core.windows.net/DimCustomers")
    dlt_obj.alias("trg").merge(df_final.alias("src"),"trg.DimCustomersKey = src.DimCustomersKey")\
        .whenMatchedUpdateAll()\
        .whenNotMatchedInsertAll()\
        .execute()
else:
    df_final.write.mode("overwrite")\
        .option("path", "abfss://gold@databrickssabharath.dfs.core.windows.net/DimCustomers")\
        .saveAsTable("databricks_cata.gold.DimCustomers")
    
```

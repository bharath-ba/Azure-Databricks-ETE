# Fact ORDERS

### #Data Reading

```python
df = spark.sql("select * from databricks_cata.silver.orders_silvers")
df.display()
```

```python
%sql
select * from databricks_cata.gold.dimproducts
```

```python
df_dimcus = spark.sql("select DimCustomersKey, customer_id as dim_customer_id from databricks_cata.gold.dimcustomers") 

df_dimpro = spark.sql("select product_id as DimProductKey, product_id as dim_product_id from databricks_cata.gold.dimproducts") 

```

Fact DataFrame

```python
df_fact = df.join(df_dimcus,df["customer_id"] == df_dimcus["dim_customer_id"],how="left").join(df_dimpro,df["product_id"] == df_dimpro["dim_product_id"],how="left")
df_fact_new = df_fact.drop('dim_customer_id','dim_product_id','customer_id','product_id')
```

```python
df_fact_new.display()
```

Upsert on Fact Table

```python
from delta.tables import DeltaTable
```

```python
if spark.catalog.tableExists("databricks_cata.gold.FactOrders"):
    dlt_obj = DeltaTable.forName(spark, "databricks_cata.gold.FactOrders")
    dlt_obj.alias("trg").merge(
        df_fact_new.alias("src"),
        "trg.order_id = src.order_id and trg.DimCustomersKey = src.DimCustomersKey and trg.DimProductKey = src.DimProductKey"
    ).whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
else:
    df_fact_new.write.format("delta") \
        .option("path", "abfss://gold@databrickssabharath.dfs.core.windows.net/FactOrders") \
        .saveAsTable("databricks_cata.gold.FactOrders")
```
